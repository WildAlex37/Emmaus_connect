# Création d'une colonne pour catégoriser les âges des répondants

```python
# Définition des limites d'âge pour chaque catégorie
age_categories = [(0, 17), (18, 29), (30, 39), (40, 49), (50, 64), (65, 74), (75, float('inf'))]
category_labels = ['Moins de 18 ans', '18-29 ans', '30-39 ans', '40-49 ans', '50-64 ans', '65-74 ans', '75 ans ou plus']
def map_age_category(age):
    for i, (lower, upper) in enumerate(age_categories):
        if lower <= age <= upper:
            return category_labels[i]
    return 'Inconnu'
df['catégorie d\'âge'] = df['Age répondant'].map(map_age_category)
df
```

```python
# Supprimer les lignes avec des valeurs NaN dans la colonne "Ouvert aux nouvelles techno"
df2_cleaned = df2.dropna(subset=["Ouvert aux nouvelles techno"])

# Calculer le pourcentage des valeurs non nulles par catégorie d'âge
open_tech_percentage = df2_cleaned.groupby(["catégorie d'âge", "Ouvert aux nouvelles techno"]).size().unstack()

# Réorganiser les colonnes dans l'ordre souhaité
ordre_ouverture = ['Oui, immédiatement', 'Oui, progressivement', 'Non', 'Ne sait pas']
open_tech_percentage = open_tech_percentage[ordre_ouverture]

# Calculer le pourcentage de chaque catégorie d'ouverture aux nouvelles technologies pour chaque tranche d'âge
open_tech_percentage_pct = open_tech_percentage.apply(lambda row: row / row.sum() * 100, axis=1)

# Créer une liste de couleurs pour les différentes catégories d'âge
colors = ['green', 'lightgreen', 'red', 'black']

# Créer le graphique à barres cumulées
plt.figure(figsize=(10, 6))
open_tech_percentage_pct.plot(kind='bar', stacked=True, color=colors)

# Ajouter des labels et un titre
plt.xlabel("Tranche d'âge")
plt.ylabel("Pourcentage des répondants")
plt.title("Vous sentez-vous prêt à adopter de nouvelles technologies ou de nouveaux services numériques ?\n\n")

# Afficher la légende à côté du graphique
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

# Afficher le graphique
plt.show()
```

# Compte des manquants ou nuls.

```python
for col in df.columns:
    # Compter le nombre de valeurs nulles et manquantes dans la colonne
    n_missing = df[col].isna().sum() + (df[col] == 'NaN').sum()
    # Afficher le nombre de valeurs manquantes pour cette colonne
    print(f"Colonne {col}: {n_missing} lignes manquantes ou nulles")

n_rows = df.shape[0]
print(f"Nombre total de lignes: {n_rows}")
```

```python
for col in df.columns:
    # Compter le nombre de valeurs nulles et manquantes dans la colonne
    n_missing = df[col].isna().sum() + (df[col] == 'NaN').sum()
    # Calculer le pourcentage de valeurs manquantes pour cette colonne
    percent_missing = (n_missing / n_rows) * 100
    # Afficher le nombre et le pourcentage de valeurs manquantes pour cette colonne
    print(f"Colonne {col}: {n_missing} lignes manquantes ou nulles ({percent_missing:.2f}%)")

    n_rows = df.shape[0]
print(f"Nombre total de lignes: {n_rows}")
```

```python
#Créer un dictionnaire pour stocker les données
data = {'Colonne': [], 'Pourcentage de valeurs manquantes': []}

#Calculer le pourcentage de valeurs manquantes par colonne
for col in df.columns:
    # Compter le nombre de valeurs nulles et manquantes dans la colonne
    n_missing = df[col].isna().sum() + (df[col] == 'NaN').sum()
    # Calculer le pourcentage de valeurs manquantes pour cette colonne
    percent_missing = (n_missing / n_rows) * 100
    # Ajouter les données au dictionnaire
    data['Colonne'].append(col)
    data['Pourcentage de valeurs manquantes'].append(percent_missing)

#Créer le DataFrame à partir du dictionnaire
result_df = pd.DataFrame(data)

#Afficher le DataFrame
result_df.sort_values(by='Pourcentage de valeurs manquantes', ascending=False).head(30)
```
